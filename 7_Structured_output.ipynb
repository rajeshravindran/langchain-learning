# STRUCTURED OUTPUT 



```python
import utility as ut

OPENAI_API_KEY = ut.OPENAI_API_KEY
LANGSMITH_API_KEY = ut.LANGSMITH_API_KEY
MODEL_NAME = ut.MODEL_NAME
```

```python
from typing_extensions import TypedDict

from langchain.agents import create_agent

class ContactInfo(TypedDict):
    name: str
    email: str
    phone: str

agent = create_agent(model=MODEL_NAME, response_format=ContactInfo)

recorded_conversation = """
we talked with John Doe. he works over at example. His number is, let's see five, five, five, one two three four five, six seven.
did you get that 
and his email was john at some email.com. he wanted to order 50 boxes of something
"""

result = agent.invoke(
    {
        "messages":[
            {
                "role":"user",
                "content":recorded_conversation
            }
        ]
    }
)

result["structured_response"]
```

# MULTIPLE DATA TYPES ARE SUPPORTED 

- PYDATIC `BaseModel`
- `TypedDict`
- `dataclasses`
- json schema (dict)

```python
from langchain.agents import create_agent

from pydantic import BaseModel

class ContactInfo(BaseModel):
    name: str
    email: str
    phone: str
    order: str

agent = create_agent(model=MODEL_NAME, response_format=ContactInfo)

recorded_conversation = """
we talked with John Doe. he works over at example. His number is, let's see five, five, five, one two three four five, six seven.
did you get that 
and his email was john at some email.com. he wanted to order 50 boxes of something
"""

result = agent.invoke(
    {
        "messages":[
            {
                "role":"user",
                "content":recorded_conversation
            }
        ]
    }
)

result["structured_response"]
```
